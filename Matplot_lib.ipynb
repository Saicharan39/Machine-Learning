Plots and data analysis

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,10,100)#considering 100 equispaced values between 0 to 10
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

Plotting data

sine wave

```python
plt.plot(x,y)
plt.show()
```

cosine function

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('sinewawe')
plt.show()
```

Parabola

```python
x=np.linspace(-20,20,40)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('parabolic signal')
plt.show()
```

```python
x=np.linspace(-5,5,100)
plt.plot(x,np.sin(x),'g')
plt.plot(x,np.cos(x),'r')
plt.show()
```

```python
x=np.linspace(-5,5,100)
plt.plot(x,np.sin(x),'g*')
plt.plot(x,np.cos(x),'r+')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])#simply area of rectangle
languages=['telugu','hindi','english','tamil']
people=[100, 20, 30, 40]
ax.bar(languages,people)
plt.xlabel('language')
plt.ylabel('population')
plt.show()

```

```python
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
lang=['english','tamil','telugu','hindi']
populn=[100,200,3,22]
ax.pie(people,labels=lang,autopct='%1.1f%%')
plt.show()
```

```python
x=np.linspace(-10,10,100)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z,color='g')
plt.show()
```

3d scatterplot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
x=28*np.random.random(100)
y=np.sin(x)
z=np.cos(x)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
